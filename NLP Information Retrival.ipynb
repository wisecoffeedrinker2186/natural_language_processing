```python
import pandas as pd

review_corpus = pd.read_csv('Restaurant_Reviews.csv')["Review"].to_list()
review_corpus
```

```python
def calc_len(sent):
    return len(sent)

def preprocessed(sent):
    lower_text = sent.lower()
    tokinized_words = nltk.word_tokenize(lower_text)
    punctuation = [',', '.', "\'", '-', "'s", '"', "``", "''","[","]","?",":"]
    stop_words = stopwords.words("english")
    clean_words = [i for i in tokinized_words if i not in punctuation]
    clean_words = [i for i in tokinized_words if i not in stop_words]
    return clean_words

def calc_vocab(sent):
    clean_words = preprocessed(sent)
    return len(set(clean_words))

def calc_nouns(sent):
    tagged_sent = pos_tag(preprocessed(sent))
    nouns = []
    for i,j in tagged_sent:
        if j == "NOUN" or "NN" in j:
            nouns.append(i)
#     print(f"The nouns in the text are : {nouns}")
    return nouns

    
def frequency_count(preped_text):
    word_freq = dict()
    for i  in preped_text:
        if i in word_freq.keys():
            word_freq[i] += 1
        else:
            word_freq[i] = 1
    word_freq = dict(sorted(word_freq.items(), key=lambda item: (item[1],item[0]),reverse=True))
    return word_freq


def most_used_word(sent):
    word_freq = frequency_count(preprocessed(sent))
    return list(word_freq.keys())[0]

def ttr(sent):
    return calc_vocab(sent) / calc_len(sent)

def no_bigrams(sent):
    return len(set(nltk.bigrams(preprocessed(sent))))

def no_trigrams(sent):
    return len(set(nltk.trigrams(preprocessed(sent))))
```

```python
# ["Length","Vocab","Freq Words","TTR","Bigrams","Trigrams"]
df = pd.DataFrame()
c = 0 
for sent in review_corpus:
    temp = [calc_len(sent), calc_vocab(sent), most_used_word(
        sent), ttr(sent), no_bigrams(sent), no_trigrams(sent),calc_nouns(sent)]
    df[sent]=temp
    c += 1
```
